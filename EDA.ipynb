# Feature Engineering

```python
# 6.1 Temporal Features 
# 1. Year (Extracts long-term trend)
df["Year"] = df.index.year

# 2. Month (Extracts monthly seasonality)
df["Month"] = df.index.month

# 3. Day of Week (0 = Monday, 6 = Sunday)
df["DayOfWeek"] = df.index.dayofweek

#  4. Week of Year (Extracts granular seasonal patterns)
# .isocalendar().week ensures standard ISO week numbering
df["WeekOfYear"] = df.index.isocalendar().week.astype(int)

# 5. Is Weekend (CORRECTED LINE: Using boolean mask instead of .apply())
df["IsWeekend"] = (df.index.dayofweek >= 5).astype(int)

# 6.2 Lag and Window Features 
# Shift AQI by 1 day, grouped by city (to avoid mixing city data)
df["AQI_lag_1"] = df.groupby("City")["AQI"].shift(1)

# 6.3 Interaction Feature 
# Simple interaction between two key variables
df["Temp_Humidity_Interaction"] = df["Temperature"] * df["Humidity"]

print("New Columns Added:")
print(df.tail(3).T)
```

# In-Depth Time-Series Analysis Per City

```python
# 7 - day Rolling Mean for PM2.5, grouped by city
df["PM2.5_MA_7"] = df.groupby("City")["PM2.5"].transform(lambda x : x.rolling(window = 7, min_periods = 1).mean())
```

7.1 Decomposition of Time-Series

Using the Additive or Multiplicative model, we can decompose the AQI time series for each city into its fundamental components: Trend, Seasonality, and Residuals (Noise).$$\text{Additive Model}: Y_t = T_t + S_t + R_t$$

Trend ($T_t$): The long-term direction of the data (e.g., generally getting worse over the years).

Seasonality ($S_t$): Repeating short-term cycles (e.g., high pollution every winter).

Residuals ($R_t$): The remaining, unexplained variation.

```python
# 7.1 Decomposition of Time-Series
from statsmodels.tsa.seasonal import seasonal_decompose

# Select a single city for decomposition 
city_data = df[df["City"] == "Delhi"]["AQI"].dropna()

# Decompose the time series (e.g. assuming a 365-day yearly cycle)
decomposition = seasonal_decompose(city_data, model = "additive", period = 365)

# Plot the components:
decomposition.plot()
plt.suptitle("AQI Time Series Decomposition for Delhi", y = 1.02)
plt.show()
```

7.3 Auto-Correlation and Partial Auto-Correlation (ACF and PACF)


These plots are crucial for identifying the order of AutoRegressive (AR) and Moving Average (MA) components, which are prerequisites for building models like ARIMA (AutoRegressive Integrated Moving Average).

ACF Plot: Shows the correlation between a time series and its own lagged values. Slow decay suggests a strong trend or a high-order AR process.

PACF Plot: Shows the correlation between a time series and its own lagged values after removing the influence of intermediate lags. A sharp cut-off suggests the order of the AR term.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# For a specific city and its differenced AQI (to achieve stationarity)
city_data = df[df["City"] == "Kolkata"]["AQI"].dropna()

# Apply a first difference to help achieve stationarity
differenced_aqi = city_data.diff().dropna()

fig, axes = plt.subplots(1, 2, figsize =(16, 5))
plot_acf(differenced_aqi, lags = 30, ax = axes[0], title = "ACF of Differenced AQI (Kolkata)")
plot_pacf(differenced_aqi, lags = 30, ax = axes[1], title = "PACF of Differenced AQI (Kolkata)")
plt.show()
```

7.4 Outlier and Anomaly Detection


High AQI values are often critical events (anomalies). Time-series analysis should specifically identify and potentially model these.

Methods: Z-Score on a rolling window, Isolation Forest, or IQR method on the residuals from the seasonal decomposition.

```python
# Use the residuals from the decomposition
residuals = decomposition.resid.dropna()
Q1 = residuals.quantile(0.25)
Q3 = residuals.quantile(0.75)
IQR = Q3 - Q1
# Anomalies are points outside 1.5 * IQR from Q1/Q3
anomalies = residuals[(residuals < Q1 - 1.5 * IQR) | (residuals > Q3 + 1.5 *IQR)]
print(f"Number of AQI anomalies in Delhi: {len(anomalies)}")
```
